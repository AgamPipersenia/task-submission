## SPOTIFY

```python
identifier = 'SPOT'
start_date = '2018-01-01'
end_date = '2020-01-01'
frequency = 'daily'
page_size = 503
#page size set to 503 to collect the entire data for 2 years in one attempt. Could have used "next" function to collect the data in multiple rounds.
next_page = ''
spot_data = intrinio.SecurityApi().get_security_stock_prices(identifier, start_date=start_date, end_date=end_date, frequency=frequency, page_size=page_size, next_page=next_page)
```

```python
dct1 = spot_data.to_dict()
df1 = pd.DataFrame.from_dict(dct1["stock_prices"])
df1
```

```python
period = 10
price_key = 'close'
start_date = '2018-01-01'
end_date = '2020-01-01'
page_size = 503
next_page = ''

spot_dc = intrinio.SecurityApi().get_security_price_technicals_dc(identifier, period=period, price_key=price_key, start_date=start_date, end_date=end_date, page_size=page_size, next_page=next_page)
```

```python
dct3=spot_dc.to_dict()
spot_dc=pd.DataFrame.from_dict(dct3['technicals'])
spot_dc
```

```python
period = 14
start_date = '2018-01-01'
end_date = '2020-01-01'
page_size = 503
next_page = ''

spot_adx = intrinio.SecurityApi().get_security_price_technicals_adx(identifier, period=period, start_date=start_date, end_date=end_date, page_size=page_size, next_page=next_page)
```

```python
dct = spot_adx.to_dict()
# dct["stock_prices"]
spot_adx = pd.DataFrame(dct['technicals'])
spot_adx
```

Dropping repeated columns

```python
spot=pd.concat([df1,spot_dc],axis=1)
spot=pd.concat([spot,spot_adx],axis=1)
spot
    
column_set = set()

# Create a list to keep track of columns to drop
columns_to_drop = []

# Iterate through columns
for col in spot.columns:
    # Convert the column data to a tuple to make it hashable
    col_data = tuple(spot[col])

    # Check if this column has the same data as a previous column
    if col_data in column_set:
        columns_to_drop.append(col)
    else:
        # Add the column data to the set
        column_set.add(col_data)

# Drop the duplicate columns
spot = spot.drop(columns=columns_to_drop)
spot
```

```python
spot_dc_signals = []

for i in range(len(spot)):
    if float(spot['close'][i]) > float(spot['upper_bound'][i]):
        spot_dc_signals.append('Sell')
    elif float(spot['close'][i]) < float(spot['lower_bound'][i]):
        spot_dc_signals.append('Buy')
    else:
        spot_dc_signals.append('Hold')


spot_signals=pd.DataFrame()
spot_signals['dc'] = spot_dc_signals
spot_signals
```

```python
adx_signals = []
adx_threshold = 25
di_threshold = 20

for i in range(len(spot)):
    try:
        adx_value = float(spot['adx'][i])
        di_pos_value = float(spot['di_pos'][i])
        di_neg_value = float(spot['di_neg'][i])

        if adx_value > adx_threshold and di_pos_value > di_neg_value and di_pos_value > di_threshold:
            adx_signals.append('Buy')
        elif adx_value > adx_threshold and di_neg_value > di_pos_value and di_neg_value > di_threshold:
            adx_signals.append('Sell')
        else:
            adx_signals.append('Hold')
    except ValueError:
        adx_signals.append('NaN')

# Create a new column 'adx' in your DataFrame and assign the signals
spot_signals['adx'] = adx_signals

# Now, 'adx' column contains the buy/sell signals, and 'NaN' for non-numeric values
# spot_signals['adx'].unique()
spot_signals['adx'].unique()
```

```python
spot_signals
```
