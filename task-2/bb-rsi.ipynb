```python
import numpy as np
signal_amzn=np.array([])
signal_aapl=np.array([])
signal_msft=np.array([])
l=len(amzn)
for i in range (0,l-1):
    if ((amzn.iloc[i,4]>amzn.iloc[i,23]) and (amzn.iloc[i+1,4]>amzn.iloc[i+1,23])):
        signal_amzn=np.append(signal_amzn,'sell')
    elif ((amzn.iloc[i,4]<amzn.iloc[i,23]) and (amzn.iloc[i+1,4]<amzn.iloc[i+1,23])):
        signal_amzn=np.append(signal_amzn,'buy')
    else :
        signal_amzn=np.append(signal_amzn,'Hold')
signal_amzn=np.append(signal_amzn,'Hold')
signal_amzn
```

```python
rsi1=np.array([])
for i in range (0,l):
    if (amzn.iloc[i,25]>70):
        rsi1=np.append(rsi1,'sell')
    elif (amzn.iloc[i,25]<30):
        rsi1=np.append(rsi1,'buy')
    else :
        rsi1=np.append(rsi1,'Hold')
rsi1
```

```python
amzn_signal=pd.DataFrame()
amzn_signal['amzn_signal_bb']=signal_amzn
amzn_signal['amzn_signal_rsi']=rsi1
amzn_signal.set_index(amzn['date'],inplace=True)
amzn_signal1=[]
for i in range (0,l):
  if amzn_signal.iloc[i,0]==amzn_signal.iloc[i,1]=='buy':
    amzn_signal1.append('buy')
  elif amzn_signal.iloc[i,0]==amzn_signal.iloc[i,1]=='sell':
    amzn_signal1.append('sell')
  else:
     amzn_signal1.append('Hold')
amzn_signal['final']=amzn_signal1
```
